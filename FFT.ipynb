```python
def data_func(time,amp,freq,phase):
    return amp*np.cos(2*np.pi*freq*time+phase)
fs = 1000.0
time = np.arange(0,0.1,1.0/fs)
N = len(time)
data = data_func(time,1.0,100.0,45.0*np.pi/180.0)#振幅1 位相45度 100Hz
plt.plot(time,data)
data_fft = np.fft.fft(data)
freq = np.arange(0,fs*0.5+fs/N, fs/N)
amp = np.absolute(data_fft[:0.5*N+1]*(2.0/N))
phase = np.angle(data_fft[:0.5*N+1]*180/np.pi)
plt.subplot(211)
plt.plot(freq,amp,'b-',label="abs")
plt.legend(loc='lower left')
plt.twinx()
plt.plot(freq,phase,"g-",label="label")
plt.legend(loc='upper right')
plt.grid()

cos = np.real(data_fft[:0.5*N+1]*2.0/N)
sin = np.imag(data_fft[:0.5*N+1]*(-2.0/N))
plt.subplot(212)
plt.plot(freq, cos, label='cosine')
plt.plot(freq, sin, label='sine')
plt.legend(loc='upper right')
plt.grid()
```

```python
class TimeSeries(object):
    def __init__(self,data,fs):
        self.data=data
        self.fs=fs
        self.N=len(data)
        self.DeltaF=fs/self.N
        self.NyN=np.floor(0.5*self.N)+1
        self.fft()
        self.time=np.arange(self.N)/fs
        self.frequency=np.arange(self.NyN)*self.DeltaF
    
    def fft(self):
        data_fft=np.fft.fft(self.data)
        self.fft_amp=np.zeros(self.NyN)
        self.fft_phase=np.zeros(self.NyN)
        N=self.N
        for i,_data_fft in enumerate(data_fft[:self.NyN]):
            if i==0:
                self.fft_amp[i]=np.real(_data_fft)/N
                self.fft_phase[i]=0.0
            elif i==self.N*0.5:
                self.fft_amp[i]=np.absolute(_data_fft)*(1.0/N)
                self.fft_phase[i]=0
            else:
                self.fft_amp[i]=np.absolute(_data_fft)*(2.0/N)
                self.fft_phase[i]=np.angle(_data_fft)*180.0/np.pi
                
        _exp=self.fft_amp*np.exp(-1.0j*self.fft_phase*np.pi/180.0)
        self.fft_cosine=np.real(_exp)
        self.fft_sine=np.imag(_exp)
        return self.fft_amp, self.fft_phase
```

```python
ts=TimeSeries(data,fs)
plt.subplot(311)
plt.plot(ts.time,ts.data)
plt.xlabel("time(sec)")
plt.ylabel("data")
plt.subplot(312)
plt.plot(ts.frequency,ts.fft_amp,"b-")
plt.xlabel("frequency(Hz)")
plt.ylabel("Amplitude")
plt.twinx()
plt.ylabel("phase(degree)")
plt.yticks(np.arange(-180,180,90))
plt.plot(ts.frequency,ts.fft_phase,"g-")
plt.subplot(313)
plt.xlabel("frequency(Hz)")
plt.plot(ts.frequency,ts.fft_cosine,label="cosine")
plt.plot(ts.frequency,ts.fft_sine,label="sine")
plt.legend()
plt.tight_layout()
```
