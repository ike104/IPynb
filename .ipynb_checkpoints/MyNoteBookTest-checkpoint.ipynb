```python
# グラフをインラインに描画する
%pylab inline
import numpy as np
# 日本語をグラフに表示するために
import matplotlib.font_manager
prop = matplotlib.font_manager.FontProperties(fname=r'C:\Windows\Fonts\meiryo.ttc', size=15)

x = np.arange(-10,10,0.1)# -10から10まで0.1ステップで
y = np.sin(x)
title(u"日本語のタイトル",fontproperties=prop)
plot(x,y)
show()
```
